### clean the data and analysis

```python
#look at survival rate by different dataframes
pclass1 = df[df["pclass"]==1]
pclass2 = df[df["pclass"]==2]
pclass3 = df[df["pclass"]==3]

```

```python
pclass1
```

```python
survival_rate_pclass1 = pclass1["survived"].sum()*100/pclass1["survived"].count()
survival_rate_pclass2 = pclass2["survived"].sum()*100/pclass2["survived"].count()
survival_rate_pclass3 = pclass3["survived"].sum()*100/pclass3["survived"].count()
```

```python
print(f"The survival rate for passanger class 1 is {survival_rate_pclass1:.1f}%.")
print(f"The survival rate for passanger class 2 is {survival_rate_pclass2:.1f}%.")
print(f"The survival rate for passanger class 3 is {survival_rate_pclass3:.1f}%.")
```
